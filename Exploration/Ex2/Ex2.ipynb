```python
# ë°ì´í„° ì „ì²˜ë¦¬ í•¨ìˆ˜
def preprocess_sentence(sentence, remove_stopwords=True):
    sentence = sentence.lower() # í…ìŠ¤íŠ¸ ì†Œë¬¸ìí™”
    sentence = BeautifulSoup(sentence, "lxml").text # <br />, <a href = ...> ë“±ì˜ html íƒœê·¸ ì œê±°
    sentence = re.sub(r'\([^)]*\)', '', sentence) # ê´„í˜¸ë¡œ ë‹«íŒ ë¬¸ìì—´ (...) ì œê±° Ex) my husband (and myself!) for => my husband for
    sentence = re.sub('"','', sentence) # ìŒë”°ì˜´í‘œ " ì œê±°
    sentence = ' '.join([contractions[t] if t in contractions else t for t in sentence.split(" ")]) # ì•½ì–´ ì •ê·œí™”
    sentence = re.sub(r"'s\b","", sentence) # ì†Œìœ ê²© ì œê±°. Ex) roland's -> roland
    sentence = re.sub("[^a-zA-Z]", " ", sentence) # ì˜ì–´ ì™¸ ë¬¸ì(ìˆ«ì, íŠ¹ìˆ˜ë¬¸ì ë“±) ê³µë°±ìœ¼ë¡œ ë³€í™˜
    sentence = re.sub('[m]{2,}', 'mm', sentence) # mì´ 3ê°œ ì´ìƒì´ë©´ 2ê°œë¡œ ë³€ê²½. Ex) ummmmmmm yeah -> umm yeah

    # ë¶ˆìš©ì–´ ì œê±° (Text)
    if remove_stopwords:
        tokens = ' '.join(word for word in sentence.split() if not word in stopwords.words('english') if len(word) > 1)
    # ë¶ˆìš©ì–´ ë¯¸ì œê±° (Summary)
    else:
        tokens = ' '.join(word for word in sentence.split() if len(word) > 1)
    return tokens
print('ì™„ë£Œ')
```

```python
# í¬ê·€ ë‹¨ì–´ì˜ í†µê³„ íŒŒì•… (threshold ë¯¸ë§Œ ë“±ì¥ ë‹¨ì–´ì˜ ê°œìˆ˜, ë¹„ìœ¨ ë“± ê³„ì‚°)
threshold = 7
total_cnt = len(src_tokenizer.word_index) # ë‹¨ì–´ì˜ ìˆ˜
rare_cnt = 0 # ë“±ì¥ ë¹ˆë„ìˆ˜ê°€ thresholdë³´ë‹¤ ì‘ì€ ë‹¨ì–´ì˜ ê°œìˆ˜ë¥¼ ì¹´ìš´íŠ¸
total_freq = 0 # í›ˆë ¨ ë°ì´í„°ì˜ ì „ì²´ ë‹¨ì–´ ë¹ˆë„ìˆ˜ ì´ í•©
rare_freq = 0 # ë“±ì¥ ë¹ˆë„ìˆ˜ê°€ thresholdë³´ë‹¤ ì‘ì€ ë‹¨ì–´ì˜ ë“±ì¥ ë¹ˆë„ìˆ˜ì˜ ì´ í•©

# ë‹¨ì–´ì™€ ë¹ˆë„ìˆ˜ì˜ ìŒ(pair)ì„ keyì™€ valueë¡œ ë°›ëŠ”ë‹¤.
for key, value in src_tokenizer.word_counts.items():
    total_freq = total_freq + value

    # ë‹¨ì–´ì˜ ë“±ì¥ ë¹ˆë„ìˆ˜ê°€ thresholdë³´ë‹¤ ì‘ìœ¼ë©´
    if(value < threshold):
        rare_cnt = rare_cnt + 1
        rare_freq = rare_freq + value

print('ë‹¨ì–´ ì§‘í•©(vocabulary)ì˜ í¬ê¸° :', total_cnt)
print('ë“±ì¥ ë¹ˆë„ê°€ %së²ˆ ì´í•˜ì¸ í¬ê·€ ë‹¨ì–´ì˜ ìˆ˜: %s'%(threshold - 1, rare_cnt))
print('ë‹¨ì–´ ì§‘í•©ì—ì„œ í¬ê·€ ë‹¨ì–´ë¥¼ ì œì™¸ì‹œí‚¬ ê²½ìš°ì˜ ë‹¨ì–´ ì§‘í•©ì˜ í¬ê¸° %s'%(total_cnt - rare_cnt))
print("ë‹¨ì–´ ì§‘í•©ì—ì„œ í¬ê·€ ë‹¨ì–´ì˜ ë¹„ìœ¨:", (rare_cnt / total_cnt)*100)
print("ì „ì²´ ë“±ì¥ ë¹ˆë„ì—ì„œ í¬ê·€ ë‹¨ì–´ ë“±ì¥ ë¹ˆë„ ë¹„ìœ¨:", (rare_freq / total_freq)*100)
```

```python
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense, Concatenate, TimeDistributed
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint


# ì¸ì½”ë” ì„¤ê³„ ì‹œì‘
embedding_dim = 128
hidden_size = 256

# ì¸ì½”ë”
encoder_inputs = Input(shape=(text_max_len,))

# ì¸ì½”ë”ì˜ ì„ë² ë”© ì¸µ
enc_emb = Embedding(src_vocab, embedding_dim)(encoder_inputs)

# ì¸ì½”ë”ì˜ LSTM 1
# encoder_lstm1 = LSTM(hidden_size, return_sequences=True, return_state=True ,dropout = 0.4, recurrent_dropout = 0.4)
encoder_lstm1 = LSTM(hidden_size, return_sequences=True, return_state=True ,dropout = 0.4)
encoder_output1, state_h1, state_c1 = encoder_lstm1(enc_emb)

# ì¸ì½”ë”ì˜ LSTM 2
encoder_lstm2 = LSTM(hidden_size, return_sequences=True, return_state=True, dropout=0.4)
encoder_output2, state_h2, state_c2 = encoder_lstm2(encoder_output1)

# ì¸ì½”ë”ì˜ LSTM 3
encoder_lstm3 = LSTM(hidden_size, return_sequences=True, return_state=True, dropout=0.4)
encoder_outputs, state_h, state_c = encoder_lstm3(encoder_output2)
```

```python
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense, Concatenate, Attention
from tensorflow.keras.models import Model

# ë””ì½”ë” ì…ë ¥ ì •ì˜
decoder_inputs = Input(shape=(None,))

# ë””ì½”ë” ì„ë² ë”©
dec_emb_layer = Embedding(input_dim=tar_vocab, output_dim=128)
dec_emb = dec_emb_layer(decoder_inputs)

# ë””ì½”ë” LSTM (ë‹¨ì¸µ êµ¬ì¡°)
decoder_lstm = LSTM(256, return_sequences=True, return_state=True, dropout=0.4)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=[state_h, state_c])

# ì–´í…ì…˜ ë©”ì»¤ë‹ˆì¦˜
attention = Attention()
attention_output = attention([decoder_outputs, encoder_outputs])  # query=decoder, value=encoder

# ì–´í…ì…˜ ì¶œë ¥ê³¼ ë””ì½”ë” ì¶œë ¥ ê²°í•©
decoder_concat_input = Concatenate(axis=-1)([decoder_outputs, attention_output])

# ì¶œë ¥ì¸µ
decoder_dense = Dense(tar_vocab, activation='softmax')
decoder_outputs = decoder_dense(decoder_concat_input)

# ìµœì¢… ëª¨ë¸
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# ìš”ì•½ ì¶œë ¥
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy')
es = EarlyStopping(monitor='val_loss', patience=2, verbose=1)

history = model.fit(
    x=[encoder_input_train, decoder_input_train],
    y=decoder_target_train,
    validation_data=([encoder_input_test, decoder_input_test], decoder_target_test),
    batch_size=256,
    callbacks=[es],
    epochs=50
)
```

### **í›ˆë ¨ëœ ëª¨ë¸ì„ ì´ì œ ì‹¤ì œ ì‚¬ìš© ê°€ëŠ¥í•˜ê²Œ ì¸ì½”ë”ì™€ ë””ì½”ë”ë¡œ ë¶„ë¦¬í•˜ê¸°**

```python
# ì›ë¬¸ í…ìŠ¤íŠ¸ ì¸ë±ìŠ¤ë¥¼ ë‹¤ì‹œ ë‹¨ì–´ë¡œ ë³€í™˜í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ (ì •ìˆ˜ â†’ ë‹¨ì–´)
src_index_to_word = src_tokenizer.index_word

# ìš”ì•½ë¬¸ í…ìŠ¤íŠ¸ë¥¼ ì •ìˆ˜ë¡œ ë³€í™˜í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ (ë‹¨ì–´ â†’ ì •ìˆ˜)
tar_word_to_index = tar_tokenizer.word_index

# ìš”ì•½ë¬¸ í…ìŠ¤íŠ¸ ì¸ë±ìŠ¤ë¥¼ ë‹¤ì‹œ ë‹¨ì–´ë¡œ ë³€í™˜í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ (ì •ìˆ˜ â†’ ë‹¨ì–´)
tar_index_to_word = tar_tokenizer.index_word

# ì¸í¼ëŸ°ìŠ¤(ì¶”ë¡ ) ë‹¨ê³„ì—ì„œ ì‚¬ìš©í•  ë‹¨ì–´ ì¸ë±ìŠ¤ ë§¤í•‘ ì¤€ë¹„ ì™„ë£Œ
print('ì™„ë£Œ')
```

```python
# ì¸ì½”ë” ì„¤ê³„
encoder_model = Model(inputs=encoder_inputs, outputs=[encoder_outputs, state_h, state_c])

# ì´ì „ ì‹œì ì˜ ìƒíƒœë“¤ì„ ì €ì¥í•˜ëŠ” í…ì„œ
decoder_state_input_h = Input(shape=(hidden_size,))
decoder_state_input_c = Input(shape=(hidden_size,))

dec_emb2 = dec_emb_layer(decoder_inputs)

# ë¬¸ì¥ì˜ ë‹¤ìŒ ë‹¨ì–´ë¥¼ ì˜ˆì¸¡í•˜ê¸° ìœ„í•´ì„œ ì´ˆê¸° ìƒíƒœ(initial_state)ë¥¼ ì´ì „ ì‹œì ì˜ ìƒíƒœë¡œ ì‚¬ìš©. ì´ëŠ” ë’¤ì˜ í•¨ìˆ˜ decode_sequence()ì— êµ¬í˜„
# í›ˆë ¨ ê³¼ì •ì—ì„œì™€ ë‹¬ë¦¬ LSTMì˜ ë¦¬í„´í•˜ëŠ” ì€ë‹‰ ìƒíƒœì™€ ì…€ ìƒíƒœì¸ state_hì™€ state_cë¥¼ ë²„ë¦¬ì§€ ì•ŠìŒ.
decoder_outputs2, state_h2, state_c2 = decoder_lstm(dec_emb2, initial_state=[decoder_state_input_h, decoder_state_input_c])
```

```python
# ì–´í…ì…˜ ë° ì†Œí”„íŠ¸ë§¥ìŠ¤ ë ˆì´ì–´ëŠ” í›ˆë ¨ ì‹œ ì •ì˜í•œ ì´ë¦„ìœ¼ë¡œ í†µì¼í•˜ëŠ” ê²ƒì´ ì¤‘ìš”í•¨ -> ë³€ìˆ˜ëª… ê·¸ëŒ€ë¡œ ê°€ì ¸ì˜¤ê¸° â—ï¸â—ï¸


# 1. ì¸í¼ëŸ°ìŠ¤ìš© ì…ë ¥ ì •ì˜
decoder_state_input_h = Input(shape=(hidden_size,))
decoder_state_input_c = Input(shape=(hidden_size,))
decoder_hidden_state_input = Input(shape=(text_max_len, hidden_size))

# 2. ë””ì½”ë” ì…ë ¥ ì •ì˜
decoder_inputs = Input(shape=(None,))
dec_emb2 = dec_emb_layer(decoder_inputs)  # í›ˆë ¨ ì‹œ ì‚¬ìš©í•œ ì„ë² ë”© ë ˆì´ì–´ ì¬ì‚¬ìš©

# 3. ë””ì½”ë” LSTM
decoder_outputs2, state_h2, state_c2 = decoder_lstm(
    dec_emb2, initial_state=[decoder_state_input_h, decoder_state_input_c]
)

# 4. ì–´í…ì…˜ ì ìš© (í›ˆë ¨ ì‹œ ì •ì˜í•œ 'attention' ê°ì²´ ì¬ì‚¬ìš©)
attn_out_inf = attention([decoder_outputs2, decoder_hidden_state_input])
decoder_inf_concat = Concatenate(axis=-1)([decoder_outputs2, attn_out_inf])

# 5. ì†Œí”„íŠ¸ë§¥ìŠ¤ ì¶œë ¥ì¸µ (í›ˆë ¨ ì‹œ ì‚¬ìš©í•œ Dense ë ˆì´ì–´ ì¬ì‚¬ìš©)
decoder_outputs2 = decoder_dense(decoder_inf_concat)

# 6. ë””ì½”ë” ì¸í¼ëŸ°ìŠ¤ ëª¨ë¸ ì •ì˜
decoder_model = Model(
    [decoder_inputs, decoder_hidden_state_input, decoder_state_input_h, decoder_state_input_c],
    [decoder_outputs2, state_h2, state_c2]
)
```

```python
import numpy as np

# ë””ì½”ë”© í•¨ìˆ˜: ì¸í¼ëŸ°ìŠ¤ ë‹¨ê³„ì—ì„œ ìš”ì•½ì„ ìƒì„±í•¨
def decode_sequence(input_seq):
    # ì¸ì½”ë” ëª¨ë¸ë¡œë¶€í„° ì¸ì½”ë” ì¶œë ¥ ë° ìµœì¢… ìƒíƒœê°’ë“¤ì„ ì–»ìŒ
    encoder_outs, state_h, state_c = encoder_model.predict(input_seq)

    # ë””ì½”ë” ì…ë ¥ì˜ ì‹œì‘ì€ í•­ìƒ <sostoken>
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = tar_word_to_index['sostoken']

    stop_condition = False
    decoded_sentence = ''

    while not stop_condition:
        # ë””ì½”ë” ëª¨ë¸ë¡œë¶€í„° ë‹¤ìŒ ë‹¨ì–´ ì˜ˆì¸¡
        output_tokens, h, c = decoder_model.predict(
            [target_seq, encoder_outs, state_h, state_c]
        )

        # ì˜ˆì¸¡ëœ ë‹¨ì–´ ì¤‘ í™•ë¥ ì´ ê°€ì¥ ë†’ì€ ë‹¨ì–´ë¥¼ ì„ íƒ
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_token = tar_index_to_word.get(sampled_token_index, '')

        # ì¢…ë£Œ í† í°ì´ ì•„ë‹ˆë¼ë©´ ê²°ê³¼ì— ì¶”ê°€
        if sampled_token != 'eostoken':
            decoded_sentence += ' ' + sampled_token

        # ì¢…ë£Œ ì¡°ê±´: <eostoken> ì´ê±°ë‚˜ ìµœëŒ€ ê¸¸ì´ ë„ë‹¬ ì‹œ
        if (sampled_token == 'eostoken') or (len(decoded_sentence.split()) >= (headlines_max_len - 1)):
            stop_condition = True

        # ë‹¤ìŒ ì…ë ¥ ì‹œí€€ìŠ¤ë¥¼ ì—…ë°ì´íŠ¸
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        # ìƒíƒœ ì—…ë°ì´íŠ¸
        state_h, state_c = h, c

    return decoded_sentence.strip()
```

## Step 4. ì‹¤ì œ ê²°ê³¼ì™€ ìš”ì•½ë¬¸ ë¹„êµí•˜ê¸° (ì¶”ìƒì  ìš”ì•½)

ì›ë˜ì˜ ìš”ì•½ë¬¸(headlines ì—´)ê³¼ í•™ìŠµì„ í†µí•´ ì–»ì€ ì¶”ìƒì  ìš”ì•½ì˜ ê²°ê³¼ë¥¼ ë¹„êµí•´ ë³´ì„¸ìš”

```python
# ì›ë¬¸ì˜ ì •ìˆ˜ ì‹œí€€ìŠ¤ë¥¼ í…ìŠ¤íŠ¸ ì‹œí€€ìŠ¤ë¡œ ë³€í™˜
def seq2text(input_seq):
    temp=''
    for i in input_seq:
        if (i!=0):
            temp = temp + src_index_to_word[i]+' '
    return temp

# ìš”ì•½ë¬¸ì˜ ì •ìˆ˜ ì‹œí€€ìŠ¤ë¥¼ í…ìŠ¤íŠ¸ ì‹œí€€ìŠ¤ë¡œ ë³€í™˜
def seq2summary(input_seq):
    temp = ''
    for i in input_seq:
        if (i!=0 and i!=tar_word_to_index['sostoken'] and i!=tar_word_to_index['eostoken']):
            temp = temp + tar_index_to_word[i] + ' '
    return temp
```
