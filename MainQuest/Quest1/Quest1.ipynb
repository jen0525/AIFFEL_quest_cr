(입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.)


```python
import random as r

#계좌 개설 함수
def create_account():
    while True:
        holder = input("예금주 이름: ")
        while True:
            try:
                balance = int(input("초기 잔액: "))
                if balance <= 0:
                    print("0보다 큰 금액을 입력해주세요.")
                    continue
                break
            except ValueError:
                print("숫자를 입력해주세요.")
        return Account(holder, balance)

#Acccount Class
class Account:
    account_count = 0

#생성자 (Magic Method)
    def __init__(self, holder, balance):
        self.holder = holder
        self.balance = balance
        self.bank_name = "SC은행"
        self.account = self.randomly_generated_number()
        self.deposit_count = 0
        self.deposit_history_list = []
        self.withdraw_history_list = []
        Account.account_count += 1

#계좌번호 랜덤 생성
    def randomly_generated_number(self):
        part1 = r.randint(100, 999)
        part2 = r.randint(10, 99)
        part3 = r.randint(100000, 999999)
        return f"{part1}-{part2}-{part3}"

#클래스 변수 출력
    @classmethod
    def get_account_num(cls):
        print(f"현재 생성된 계좌의 개수는 {cls.account_count}개입니다.")

#입금 함수
    def deposit(self, amount):
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return
        self.balance += amount
        self.deposit_count += 1
        self.deposit_history_list.append(amount)
        print(f"{amount}원이 입금되었습니다. 현재 잔액: {self.balance:,}원")
        if self.deposit_count == 5:
            self.add_interest()

#출금 함수
    def withdraw(self, amount):
        if amount < 1:
            print("출금은 최소 1원 이상만 가능합니다.")
            return
        if amount > self.balance:
            print("출금 금액이 계좌 잔고를 초과할 수 없습니다.")
            return
        self.balance -= amount
        self.withdraw_history_list.append(amount)
        print(f"{amount}원이 출금되었습니다. 현재 잔액: {self.balance:,}원")

#정보 출력 메소드
    def display_info(self):
        formatted_balance = f"{self.balance:,}"
        print(f"은행이름: {self.bank_name}, 예금주: {self.holder}, "
              f"계좌번호: {self.account}, 잔고: {formatted_balance}원")

#이자 함수
    def add_interest(self):
        interest = int(self.balance * 0.01)
        self.balance += interest
        self.deposit_count = 0
        print(f"이자가 {interest:,}원 지급되었습니다. 현재 잔액: {self.balance:,}원")

    def deposit_history(self):
        print(f"{self.holder}님의 입금 내역: {self.deposit_history_list}")

    def withdraw_history(self):
        print(f"{self.holder}님의 출금 내역: {self.withdraw_history_list}")

#입금 계좌 찾기
def deposit_money(account_list):
    acc_num = input("입금할 계좌번호: ")
    try:
        amount = int(input("입금액: "))
    except ValueError:
        print("잘못된 입력입니다.")
        return
    for acc in account_list:
        if acc.account == acc_num:
            acc.deposit(amount)
            return
    else:
        print("계좌를 찾을 수 없습니다.")

#출금 계좌 찾기
def withdraw_money(account_list):
    acc_num = input("출금할 계좌번호: ")
    try:
        amount = int(input("출금액: "))
    except ValueError:
        print("잘못된 입력입니다.")
        return
    for acc in account_list:
        if acc.account == acc_num:
            acc.withdraw(amount)
            return
    else:
        print("계좌를 찾을 수 없습니다.")

# 잔고 100만원 이상 고객 출력 함수
def display_high_balance_accounts(account_list):
    print("\n[잔고 100만원 이상 고객 정보]")
    for account in account_list:
        if account.balance >= 1000000:
            account.display_info()

# 입출금 내역 확인 함수
def display_deposit_and_withdraw_history(account_list):
    account_number = input("내역을 확인할 계좌 번호를 입력하세요: ")
    for account in account_list:
        if account.account == account_number:
            account.deposit_history()
            account.withdraw_history()
            return
    print("입력한 계좌 번호를 찾을 수 없습니다.")

account_list = []

```
